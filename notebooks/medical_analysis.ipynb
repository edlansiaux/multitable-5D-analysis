# Analyse Médicale 5D (MIMIC-IV Demo)

Ce notebook démontre l'utilisation du framework **RHT** sur des données médicales simulées, illustrant les 5 dimensions de complexité :
1. Volume
2. Variables Multiples
3. Haute Cardinalité (Codes ICD)
4. Relations (Patients-Admissions-Soins)
5. Temporalité (Signes vitaux)

```python
import sys
import os
sys.path.append('..') # Ajout de la racine du projet

from mt5d.datasets.synthetic import SyntheticMultiTableGenerator
from mt5d.core.pipeline.mt5d_pipeline import MT5DPipeline
from mt5d.utils.visualization import plot_relational_hypergraph
```

## 1. Génération de Données

```python
generator = SyntheticMultiTableGenerator(num_patients=200)
tables, relationships = generator.generate()

print("Tables générées :", list(tables.keys()))
print("Exemple (Patients) :")
tables['patients'].head()
```

## 2. Exécution du Pipeline 5D

```python
pipeline = MT5DPipeline(config_path='../configs/default_config.yaml')
results = pipeline.run(tables, relationships, target_task='mortality')
```

## 3. Visualisation de l'Hypergraphe

```python
g = results['hypergraph']
plot_relational_hypergraph(g, title="Graphe Relationnel Médical (Fragment)")
```
